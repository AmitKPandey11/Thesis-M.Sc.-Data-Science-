```python
def cleanTxt(text):
    text=re.sub('@[A-Za-z0-9]+',"",text)
    text=re.sub('/d+',"",text)
    text=re.sub("b'","",text)
    text=re.sub('b"',"",text)
    text=re.sub('RT',"",text)#retweeted tweet
    text=re.sub('TW',"",text)
    text=re.sub('-',"",text)
    text=re.sub('&amp;',"",text)
    text=re.sub(':',"",text)
    text=re.sub('https?:\/\/\S+','',text)
    text=re.sub('https?.\/\/\S+','',text)
    text=re.sub('#',"",text)
    text=text.rstrip('\"')
    text=text.rstrip("\'")
    text=re.sub('\\\\',"",text)
    text=re.sub('\\\\\w{1,3}',"",text)
    return text
        
    
```

```python
def random_colours(number_of_colors):
    '''
    Simple function for random colours generation.
    Input:
        number_of_colors - integer value indicating the number of colours which are going to be generated.
    Output:
        Color in the following format: ['#E86DA4'] .
    '''
    colors = []
    for i in range(number_of_colors):
        colors.append("#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]))
    return colors
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
```

```python
def stemma(tw):

    stemmer = SnowballStemmer("english") #SnowballStemmer("english", ignore_stopwords=True)

    count = 0
    stemmed = []
    for sentence in tw:
        word_tokens = word_tokenize(sentence)
        for word in word_tokens:
            stemmed.append(stemmer.stem(word))
        sentence = ' '.join(stemmed)
        tw.iloc[count] = sentence
        count+=1
        stemmed = []
    return(tw)
```

```python
def lemma(tw):

    lmtzr = WordNetLemmatizer()

    count = 0
    stemmed = []
    for sentence in tw:    
        word_tokens = word_tokenize(sentence)
        for word in word_tokens:
            stemmed.append(lmtzr.lemmatize(word))
        sentence = ' '.join(stemmed)
        tw.iloc[count] = sentence
        count+=1
        stemmed = []
    return(tw)
```

```python
tw['Tweets_cleaned'] = lemma(tw['Tweets_cleaned'])
tw['Tweets_cleaned'] = stemma(tw['Tweets_cleaned'])
```
